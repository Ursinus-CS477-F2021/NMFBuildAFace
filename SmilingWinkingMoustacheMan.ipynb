```python
import numpy as np
import matplotlib.pyplot as plt
import skimage.io
```

```python
W = []
plt.figure(figsize=(10, 10))
shape = ()
for i in range(9):
    img = skimage.io.imread("{}.png".format(i))
    img = np.sum(img, axis=2) # Only need grayscale
    img[img > 0] = 1.0
    shape = img.shape
    W.append(img.flatten())
    plt.subplot(3, 3, i+1)
    plt.imshow(img, cmap='gray')
    plt.colorbar()
    plt.title("Column {}".format(i))
W = np.array(W).T
plt.savefig("FaceColumns.png", bbox_inches='tight', facecolor='white')
```
